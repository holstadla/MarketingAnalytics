Qualitative reasoning for quick insights:

```
#total revenue, cost, quantities sold, and profit for each country over data timespan 
grouped_sales = sales.groupby('Retailer country')['Revenue', 'Planned revenue', 'Product cost', 'Quantity', 'Gross profit'].sum()
grouped_sales
```

```
grouped_sales.sort_values('Quantity')
```

Denmark has the lowest number of sales; US has highest

```
grouped_sales.sort_values('Gross profit')
```

Denmark has lowest gross profit; US has highest. Quanity sold correlates to gross profit.

```
grouped_mean_sales = sales.groupby('Retailer country')['Revenue', 'Planned revenue', 'Product cost', 'Quantity', 'Gross profit', 'Unit sale price'].mean()
grouped_mean_sales
```

```
grouped_mean_sales.sort_values('Revenue')
```

Sweden has lowest revenue. Top 3 revenue goes to US, China, and UK. Average sale price among all countries sanges from around 42 - 49 USD. 

```
#group by retailer country to see what countries were affected the worst when sales dipped
#order by lowest values
sales.dropna().groupby('Retailer country')['Revenue','Planned revenue','Product cost','Quantity','Unit cost','Unit price','Gross profit','Unit sale price'].min()
```

Brazil, Finland, and Spain have the highest lowest gross profit. They are among a few countries that have not had a negative cumulative profit over the years. 

```
#Find the totals for categories
sales.groupby('Year')['Revenue','Planned revenue','Product cost','Quantity','Unit cost','Unit price','Gross profit','Unit sale price'].sum()
```

Revenue, quantity, and gross profit has decreased in 2007. 

```
sales.groupby('Product line')['Revenue','Planned revenue','Product cost','Quantity','Unit cost','Unit price','Gross profit','Unit sale price'].sum()
```

Can see here that camping equipment is the most lucrative in the product line.

```
sales.groupby('Product type')['Revenue','Planned revenue','Product cost','Quantity','Unit cost','Unit price','Gross profit','Unit sale price'].sum()
```

Cooking gear is the most lucrative product type. Sleeping bags have the second highest revenue while having the highest unit sale price. However, sleeping bags have the second lowest quantity sold. There is potential here to focus on more promotions to sell sleeping bags in order to increase profits. 

```
sales.groupby('Order method type')['Revenue','Planned revenue','Product cost','Quantity','Unit cost','Unit price','Gross profit','Unit sale price'].sum()
```

Online was the most lucrative method of purchasing.

```
#transform numeric unit cost into a categorical column:
import numpy as np

def cat_gen(x): 
  if pd.isnull(x): 
    return np.nan
  elif x <= 2.76: 
    return "cheap"
  elif 2.76<=x<=9.0: 
    return "medium"
  elif 9.0<=x<=34.97: 
    return "moderate"
  else: 
    return "expensive"

sales['Cost category'] = sales['Unit cost'].map(cat_gen)
sales['Cost category'].value_counts(dropna = True)
```
