```
#Read files
datafile = r'/gdrive/My Drive/Data Projects/Marketing Analytics/Lesson 2/sales.csv'
sales = pd.read_csv(datafile) #creates a dataframe
sales.head()
```

```
#plot revenue's distribution with the kde parameter
sales['Revenue'].plot(kind='kde')
```

```
#group revenue by order method type and make a barplot
sales.groupby('Order method type').sum().plot(kind='bar', y='Revenue')
```

```
#group columns by year and create boxplots to get an idea on a relative scale
sales.groupby('Year')['Revenue', 'Planned revenue', 'Quantity', 'Gross profit'].plot(kind='box')
```
